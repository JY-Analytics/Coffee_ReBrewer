```python
import pandas as pd
import numpy as np
import plotly.express as px
from dash import Dash, dcc, html, Input, Output
```

```python
df = pd.read_csv("philly_reviews_asba.csv")
df.head(2)
```

```python


fig = px.density_mapbox(df, lat = 'latitude', lon = 'longitude', z = 'composite_score',
                        radius = 7,
                        center = dict(lat = 39.99, lon = -75.1),
                        zoom = 10,
                        mapbox_style = 'open-street-map',
                        opacity = 1,
                        color_continuous_scale = 'oranges')
fig.show()
```
